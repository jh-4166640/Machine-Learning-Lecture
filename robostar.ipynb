```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np


robostar = yf.download('090360.KQ', start='2023-01-01', end='2024-12-31')
close_price = np.array(robostar['Close'], dtype=np.float32)
train_dates = robostar.index
date =  robostar.index.values
date_short = np.array([str(d)[:10] for d in date])

```

```python
close_idx = 0
high_idx = 1
low_idx = 2
open_idx = 3
volume_idx = 4
ma_idx = 5

moveDay = 30 # 초단기 이동평균선
moveAverage = np.zeros((robostar.shape[0],1))


for day in range(moveDay-1, robostar.shape[0]):
  sum = 0
  for idx in range(0,moveDay):
    sum += robostar[day-idx,close_idx]
  mA = sum / moveDay
  moveAverage[day,] = mA

rate_data = np.diff(robostar[:, :4], axis=0) / robostar[:-1, :4] * 100

data = np.concatenate((robostar, moveAverage), axis=1)   #이동 평균선 추가
data = np.concatenate((data[1:,:], rate_data), axis=1)   #변화율 추가
data = data[moveDay-2:,:]
date=date[moveDay-2:]
```

```python
from sklearn.preprocessing import MinMaxScaler
# 데이터 정규화
scaler = MinMaxScaler()
norm_data = scaler.fit_transform(data)
```

```python
# train test 분할
train_ratio = 0.7
lookback = 5 # 예측을 할 때 얼마나 넣고 할거냐

train_idx = int(norm_data.shape[0] * train_ratio)
train_data = norm_data[:train_idx]
test_data = norm_data[train_idx:]

x_train = []
t_train = []

for idx in range(train_data.shape[0] - lookback):
  x_train.append(train_data[idx:idx+lookback])
  t_train.append(train_data[idx+lookback,0:volume_idx+1].T)
x_train = np.array(x_train)
t_train = np.array(t_train)
```

```python
import keras


# model = keras.models.Sequential() #LSTM
# model.add(keras.layers.LSTM(units=50,activation='tanh',return_sequences=True,input_shape=(lookback,10)))
# model.add(keras.layers.LSTM(units=25))
# model.add(keras.layers.Dense(units=5))
# model.compile(optimizer='adam', loss='mean_squared_error')


model = keras.Sequential([ # CNN-LSTM
    keras.layers.Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(lookback, 10)),
    keras.layers.MaxPooling1D(pool_size=2),
    keras.layers.LSTM(100, activation='tanh'),
    keras.layers.Dense(5)
])
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
epoch = 300

checkpoint_cb = keras.callbacks.ModelCheckpoint(
    "best_model.h5",
    save_best_only=True,        # validation_loss 기준 최적 모델만 저장
    monitor="val_loss",
    mode="min"
)
earlystop_cb = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,       # 10 Epoch 동안 개선 없으면 종료
    restore_best_weights=True  # 종료 시 최적 가중치 복원
)

history = model.fit(x_train[:,:],t_train,epochs=epoch,batch_size=32,callbacks=[checkpoint_cb, earlystop_cb])
```

```python
from matplotlib import figure
x_test=[]

for idx in range(test_data.shape[0] - lookback):
  x_test.append(test_data[idx:idx+lookback])
x_test = np.array(x_test)

model = keras.models.load_model("best_model.h5")
y_pred_norm = model.predict(x_test[:,:])  # 예측 결과 (정규화된 값)

num_features = data.shape[1]
dummy = np.zeros((y_pred_norm.shape[0], num_features))
dummy[:, :y_pred_norm.shape[1]] = y_pred_norm
y_pred_full = scaler.inverse_transform(dummy)
y_pred = y_pred_full[:, :y_pred_norm.shape[1]]

figure=plt.figure(figsize=(12,6))
plt.plot(data[train_idx+lookback:,close_idx])
plt.plot(y_pred[:,close_idx])
plt.title("Robostar[090360.KQ] Stock predict(lookback="+str(lookback)+")",fontsize=22)
plt.legend(["Real","Predict"],fontsize=18)

plt.ylabel("price[won]",fontsize=18)
plt.xlabel("date",fontsize=18)
plt.grid(visible=True)
plt.xticks(
    [0, 100, 200, 300],
    [
        date_short[train_idx + lookback],
        date_short[train_idx + lookback + 100],
        date_short[train_idx + lookback + 200],
        date_short[train_idx + lookback + 300]
    ],
    fontsize=18
)

plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np


robostar = yf.download('090360.KQ', start='2023-01-01', end='2024-12-31')
close_price = np.array(robostar['Close'], dtype=np.float32)
train_dates = robostar.index
date =  robostar.index.values
date_short = np.array([str(d)[:10] for d in date])

test_robostarFinance = yf.download('090360.KQ', start='2025-01-01')
test_close = np.array(test_robostarFinance['Close'], dtype=np.float32)
test_dates = test_robostarFinance.index


max_price = np.max(close_price)
train_norm = close_price / max_price
test_norm = test_close / max_price

# train test 분할
lookback = 10 # 예측을 할 때 얼마나 넣고 할거냐

train_data = train_norm
test_data = test_norm

x_train = []
t_train = []
x_test = []
for idx in range(train_data.shape[0] - lookback):
  x_train.append(train_data[idx:idx+lookback])
  t_train.append(train_data[idx+lookback].T)
x_train = np.array(x_train)
t_train = np.array(t_train)

for idx in range(test_data.shape[0] - lookback):
  x_test.append(test_data[idx:idx+lookback])
x_test = np.array(x_test)

```

```python
from torch import dropout
import keras

# model = keras.models.Sequential() #LSTM
# model.add(keras.layers.LSTM(units=50,activation='tanh',return_sequences=True,input_shape=(lookback,10)))
# model.add(keras.layers.LSTM(units=25))
# model.add(keras.layers.Dense(units=5))
# model.compile(optimizer='adam', loss='mean_squared_error')
dropout= 0.1

model = keras.Sequential([ # CNN-LSTM
    keras.layers.Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(lookback, 1)),
    keras.layers.MaxPooling1D(pool_size=2),
    keras.layers.LSTM(128, activation='tanh',return_sequences=True),
    keras.layers.Dropout(rate=dropout),
    keras.layers.LSTM(128, activation='tanh',return_sequences=True),
    keras.layers.Dropout(rate=dropout),
    keras.layers.LSTM(128, activation='tanh',return_sequences=True),
    keras.layers.Dropout(rate=dropout),
    keras.layers.LSTM(128, activation='tanh',return_sequences=True),
    keras.layers.Dropout(rate=dropout),
    keras.layers.LSTM(128, activation='tanh',return_sequences=True),
    keras.layers.Dropout(rate=dropout),
    keras.layers.LSTM(128, activation='tanh'),
    keras.layers.Dropout(rate=dropout),
    keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
epoch = 300

checkpoint_cb = keras.callbacks.ModelCheckpoint(
    "best_model.h5",
    save_best_only=True,        # validation_loss 기준 최적 모델만 저장
    monitor="val_loss",
    mode="min"
)
earlystop_cb = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,       # 10 Epoch 동안 개선 없으면 종료
    restore_best_weights=True  # 종료 시 최적 가중치 복원
)

history = model.fit(x_train,t_train,epochs=epoch,batch_size=32,callbacks=[checkpoint_cb, earlystop_cb])
```

```python
model = keras.models.load_model("best_model.h5")
y_pred_norm = model.predict(x_test)  # 예측 결과 (정규화된 값)


plt.figure(figsize=(20,6))

# ---- (1) 전체 실제값 ground truth 구성 ----
# train 부분: y_norm[:,0]
# test 부분: y_test[:,0]
y_train_plot = train_data * max_price
y_test_plot  = test_data  * max_price

# 실제값 전체 연결
y_all = np.concatenate([y_train_plot, y_test_plot], axis=0)

plt.plot(y_all, label="Ground Truth (Train + Test)", linewidth=2)


# ---- (2) Pred는 test 부분만 표시 ----
# pred_norm 은 test에 대한 예측만 존재함
pred_test = y_pred_norm[:,0].reshape(-1,1) * max_price
# pred 전체 배열 생성 (train 구간은 NaN)
pred_all = np.empty_like(y_all)
pred_all[:] = np.nan
pred_all[len(y_train_plot)+10:] = pred_test  # test 구간에만 pred 삽입

plt.plot(pred_all, label="Prediction (Test Only)", linewidth=2)


# ---- (3) Train/Test 경계선 표시 ----
boundary_idx = len(y_train_plot)
plt.axvline(boundary_idx, color='k', linestyle='--', linewidth=2,
            label="Train/Test Boundary (2025-01-01)")


plt.title("LSTM Forecasting (Train + Test)")
plt.grid()
plt.legend()
plt.show()
```
